# Task 1

Inject multiple signals with different parameters and arrive to a mismatch graph 

- generate several signals with different frequencies, but keep it simple by fixing the other parameters:
    - sigma: 0.25
    - phi: 0 rad
    - A: 1
    
- probe them with template banks and probe them at their known signal to generate a mismatch, from the recovered parameters and the actual parameters

```python
A = 1
N = 100

f = np.linspace(10, 100.5, N)
phi = [0.0]
sigma = [0.25]#np.linspace(0,A,1)

source = {}
i = 0
for j, k, l in itertools.product(f, phi, sigma):
    i+=1
    source[f's{i-1}'] = {'name':f's{i-1}.txt','frequency':j, 'phase':k, 'noise':l}
    
#print(source)
```

```python
#print(source)
signals = list(source.keys())

source_files = []
for i in range(len(signals)):
    source_files.append(generate_source(outputFile=source[f's{i}']['name'],sigma=sigma, A=100
                                        ,f=source[f's{i}']['frequency']))
```

```python
def get_mismatch(datafile,tbank=None,freq=None,N_df=300, 
                f_min=0.5,f_max=100.5,mod=1):
    #entry = (datafile.split('../code/data_lab2/')[1]).split('.txt')[0]
    #print(entry)
    #print(source[entry]['frequency'])
    if tbank is None:
        tbank = '../code/data_lab1/tbank1.txt'

    results_file = prober(datafile, outputFile='u.txt',templateFile=tbank, mod=1)
    results = np.loadtxt(results_file)
    # Plot best match
    if mod == 1 :
        # For S_1 get the maximum
        best_phi, best_f, best_val = results[np.argmax(results[:,2])]
        tv_phi, tv_f, tv_val = prober(datafile, f=freq, mod=mod)[1]
        res1 = prober(datafile, f=freq, mod=mod)
        
        M_i = 1 - abs(float(best_val)/float(tv_val))        

    return M_i
```

```python
N = 20
tbank = f'../code/data_lab2/temp_bank_mismatch{N}.txt'

with open(tbank,'w') as file:
    for fr in np.linspace(0.5,100.5,N):
        file.write(f'{fr} 0\n')

mismatches = []
for i, path_tofile in enumerate(source_files):
    mismatches.append(get_mismatch(path_tofile, tbank=tbank,freq=f[i], mod=1, N_df=10000, f_min=f[0], f_max=f[-1]))

plt.figure(figsize=(15,5))
plt.xlabel('frequency [Hz]')
plt.ylabel('Mismatch')
plt.plot(f, mismatches, '-o')
```

# task 2

- generate signals:
    - random noise
    - signal + random noise
- probe them at the known frequency and phase of the injected signal:
      - one can do it with a template bank
      - or simply use prober at that frequency and phase

- Do it over and over again to generate a statistic of detection from prober output at that specific frequency and phase

- plot a histogram 

```python
# Here I use a Template bank

n_dist = 100

As = 2 
sigma = 27
freq = 10
phase = 0
mod=1

temp_bank = "../code/data_lab2/histogram_test_tb.tb"

with open(os.path.join(temp_bank),'w') as tbfile:
    for f in np.linspace(0, 15, 20):
        tbfile.write("{0} {1}\n".format(f,phase))

        
tmp_bank = temp_bank.split('../code/data_lab2/')[1]

prober_results_test = {'signal':np.zeros(n_dist), 'noise':np.zeros(n_dist)}

for i in range(n_dist):
    infile = generate_source(outputFile=f'N-source_({tmp_bank}){i}.txt',A=0,sigma=sigma, N=1000)
    resultf = prober(infile,outputFile=f'N-prob_({tmp_bank}).txt', templateFile=temp_bank, mod=mod)
    results = np.loadtxt(resultf)
    ind = np.argwhere(results[:,0]>=freq).flatten()[0]
    
    prober_results_test['noise'][i] = results[:,2][ind]

    infile = generate_source(f'S-source_({tmp_bank}){i}.txt',A=As, sigma=sigma,f=freq, N=1000)
    resultf = prober(infile,outputFile=f'S-prob_({tmp_bank}).txt', templateFile=temp_bank, mod=mod)
    results = np.loadtxt(resultf)
    ind = np.argwhere(results[:,0]>=freq).flatten()[0]

    prober_results_test['signal'][i] = results[:,2][ind]

plt.hist(prober_results_test['signal'], facecolor='steelblue', alpha =0.5,label="s+n")
plt.hist(prober_results_test['noise'], facecolor='coral', alpha=0.5, label="n")
plt.xlabel("S_{}".format(1))
plt.ylabel("Count")
plt.legend()
plt.title(f"Method 1 - Sigma {sigma}, - amplitude {As}\n probing f={freq}")
plt.show()
plt.close()

#print(prober_results_test['signal'])

```
